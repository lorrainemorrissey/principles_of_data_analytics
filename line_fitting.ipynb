## Line Fitting 

```python
# Numberical Arrays.
import numpy as np

# Plotting
import matplotlib.pyplot as plt
```

#### Mathematics

> **From ChatGPT**
>
> A **polynomial** is an expression with terms made of variables and constants, using only addition, subtraction, and multiplication, with whole-number exponents. 
>
>The **degree** of a polynomial is the highest exponent of the variable.
>
>**Example:**  
>\( 4x^3 + 2x^2 - x + 7 \) is a polynomial.  
>Its **degree** is **3** (from \( x^3 \)).
>
>



>**From ChatGPT**
>
>`numpy.polyfit` fits a polynomial of a specified degree to a set of data points by minimizing the least squares error. It returns the coefficients of the polynomial, starting from the highest degree term. Essentially, it finds the best curve (or line) that approximates the relationship between two datasets. 
>
>https://chatgpt.com/share/680e0cd4-3060-8013-9110-77e7871c5647

A straight line in two dimensions can always be written as:

y = mx + c

### Data

```python
# Create x values

x = np.linspace(0.0, 10.0, 31)

# Create y values

y = 3.0 * x
```

```python
# Crate a scatterplot

fig, ax = plt.subplots()

ax.scatter(x,y)
```

```python
# Use polyfit to fit a line to the data

m, c = np.polyfit(x, y, 1)

# Show m and c
m, c

```

```python
# Crate a scatterplot

fig, ax = plt.subplots()

ax.scatter(x, y)

ax.plot(x, m*x+c, color='red')
```

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x + 0.0 + np.random.normal(0.0, 0.3, size=x.size)
```

```python
# Crate a scatterplot

fig, ax = plt.subplots()

ax.scatter(x, y)
```

```python
# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c
```

```python
# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color='red');
```

## All Together

```python
# Create x values.
x = np.linspace(0.0, 10.0, 31)

# Create y values.
y = 3.0 * x + 1.0 + np.random.normal(0.0, 1.0, size=x.size)

# Use polyfit to fit a line to the data.
m, c = np.polyfit(x, y, 1)

# Show.
m, c

# Create a scatterplot.
fig, ax = plt.subplots()

# Plot the data.
ax.scatter(x, y)

ax.plot(x, m * x + c, color='red')

ax.set_title(f'y = {m:.2f}x + {c:.2f}');
```
