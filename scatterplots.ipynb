```python
# Example linked above 

#import matplotlib.pyplot as plt
#import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

### Pairs of Features


```python
# https://scikit-learn.org/stable/datasets/toy_dataset.html
diabetes = datasets.load_diabetes(scaled=False)

# Show keys
diabetes.keys()
```

```python
# Feature names
# 10 values 
diabetes.feature_names
```

```python
# Data.
diabetes.data
```

```python
# Blood Pressure
# Show the data from index 3 (i.e 4th column)
bp = diabetes.data[:, 3]
#[:,3] - outer dimension everything, column 3

# Show - first 10
bp[:10]
```

```python
# BMI

bmi = diabetes.data[:, 2]

# Show.
bmi[:10]
```

### Relationships

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(bmi, bp)

# Labels
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```

### Colours

```python
# Create a figure and an axis
fig, ax = plt.subplots()

# Scatter plot
ax.scatter(bmi, bp, marker='.')

# Labels
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```

```python
# Sex.
sex = diabetes.data[:, 1]

# Show first ten
sex[:10]
```

```python
# Unique sex values
np.unique(sex)
```

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

# Scatter plot
# Change colour
ax.scatter(bmi, bp, c=sex, marker='.')

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```

### Size

```python
# Age
age = diabetes.data[:, 0]

# Show first ten
age[:20]
```

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

# Scatter plot.
# Change colour.
ax.scatter(bmi, bp, c=sex, s=age, marker='^')

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```
