# Simple Linear Regression

https://en.wikipedia.org/wiki/Simple_linear_regression

>**From ChatGPT**
>
>**Linear regression** is a simple statistical method used to model the relationship between a **dependent variable** (what you want to predict) and one or more **independent variables** (the inputs).  
>
>It tries to fit a straight line through the data points that best describes how the dependent variable changes with the independent variable(s). The equation usually looks like:
\[
y = mx + b
\]
>
>where:  
>- \( y \) = predicted value,  
>- \( x \) = input value,  
>- \( m \) = slope of the line (how much \( y \) changes when \( x \) changes),  
>- \( b \) = y-intercept (value of \( y \) when \( x = 0 \)).
>
>The goal is to find the best \( m \) and \( b \) that minimize the difference between the predicted and actual values.  
>
>
>*https://chatgpt.com/share/680e40d5-2904-8013-aaf7-0e43ac133a7a*

```python
# Numerical arrays
import numpy as np

# Plotting
import matplotlib.pyplot as plt
```


Straight line can be written as:
$y = mx + c$  


$y = 2x + 1$

```python
# An array of x values
x = np.linspace(0.0, 10, 21)

# An array of y values with noise
y = 2.0 * x +1.0 + np.random.normal(0.0, 1.0, x.size)
```

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.scatter(x, y, marker='.', label='$y = 2x + 1$')

# Add legend
ax.legend();
```

```python
#Polyfit

np.polyfit(x, y, 1)
```

### Calculatiuon from wikipedia  

https://en.wikipedia.org/wiki/Simple_linear_regression#Formulation_and_computation

```python
# Mean of x and y

xbar = np.mean(x)
ybar = np.mean(y)
```

```python
# Numerator
num = np.sum((x-xbar) * (y - ybar))

# Denominator
den = np.sum((x-xbar) **2)

# Slope
m = num / den

# show
m
```

```python
# Intercept
c = ybar - (m * xbar)

# show
c
```

## Without Intercept

https://en.wikipedia.org/wiki/Simple_linear_regression#Simple_linear_regression_without_the_intercept_term_(single_regressor)  

$y = mx$

```python
# x values

x = np.linspace(0.0, 10.0, 21)

# y values with noise

y = 3.0 * x + np.random.normal(0.0, 1.0, x.size)
```

```python
# Plot
fig, ax = plt.subplots()

# Scatter
ax.scatter(x, y, marker='.',label='$y = 3x$')

# Add line
ax.plot(x, 3.0 * x,'r', label='Best fit')

# Add Legend
ax.legend();
```

```python
# Fit.
np.sum(x * y) / np.sum(x * x)
```

## Changing m and c

```python
# Comparing different m's (slopes).
fig, ax = plt.subplots()

# Different sloped lines.
ax.plot(x, -1.0 * x, label='$y = -1.0x$')
ax.plot(x, 1.0 * x, label='$y = 1.0x$')
ax.plot(x, 2.0 * x, label='$y = 2.0x$')
ax.plot(x, 3.0 * x, label='$y = 3.0x$')

ax.legend();
```

```python
# Comparing different c's (y-intercepts).
fig, ax = plt.subplots()

# Different sloped lines.
ax.plot(x, x - 1.0, label='$y = x - 1.0$')
ax.plot(x, x + 1.0, label='$y = x + 1.0$')
ax.plot(x, x + 2.0, label='$y = x + 2.0$')
ax.plot(x, x + 3.0, label='$y = x + 3.0$')

ax.legend();
```
