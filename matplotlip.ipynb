```python
# Plots
import matplotlib.pyplot as plt

# Numerical Arrays
import numpy as np
```

# Resources

## Plotting functions

```python

# X values

x = np.linspace(0.0, 10.0, 101)

x
```

```python
# y values

y = x**2

y
```

```python
# Plot x vs y
plt.plot(x, y, label='$y = x^2$');

# Add Legend
plt.legend()

# Add Labels
plt.xlabel('x')
plt.ylabel('$x^2$')

# Add Tite
plt.title('A Lovely Plot');
```

## Comparing Plots

```python
# Domain of x values.
x = np.linspace(0.0, 10.1, 101)

# x squared - polynomial 
y1 = x**2

# 2 to the power of x - exponential
y2 = 2**x
```

```python
# plot x vs y1.
plt.plot(x, y1, label='$y = x^2$')

#plotx vs y2
plt.plot(x, y2, label='$y = 2^x$')

# Add a legend
plt.legend()

# Add labels
plt.xlabel('x')
plt.ylabel('y')

# Add a title
plt.title('Two Lovely Plots');
```
