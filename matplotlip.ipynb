```python
# Plots
import matplotlib.pyplot as plt

# Numerical Arrays
import numpy as np
```

# Resources

## Plotting functions

```python

# X values

x = np.linspace(0.0, 10.0, 101)

x
```

```python
# y values

y = x**2

y
```

```python
# Plot x vs y
plt.plot(x, y, 'r:', label='y = x^2');
plt.legend()
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('A Lovely Plot');
```
